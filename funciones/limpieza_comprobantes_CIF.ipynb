```python
import pandas as pd
import numpy as np
import openpyxl
import os
```

```python
# Creamos la variable con el nombre del archivo

nombre_archivo = 'comprobantes2022.XLS'
```

```python
# Creamos la variable que contenga la ruta del archivo

ruta_archivo = f'../data/{nombre_archivo}'
```

```python
# Leemos el archivo con pandas y lo guardamos en una variable llamada file_excel que creamos, y sacamos las keys para ver las hojas que contiene el libro de excel
# 
file_excel = pd.read_excel(ruta_archivo, na_values= 'NN', sheet_name=None, skiprows=3, header = None)
file_excel.keys()
```

```python
# Creamos la variable DF, donde concatenamos todaas las hojas del archivo de excel

df0 = pd.concat(file_excel, ignore_index= True)
df0
```

```python
# Creamos una variable nueva llamada DF2 donde guardamos sobre una copia de DF, a su vez redondeamos a dos digitos los decimales:

df = df0.copy()
df = df.round(2)
df
```

```python
# Eliminamos las 3 primeras filas ya que tiene informacion que nos resulta inutil, luego reiniciamos el indice 

df2 = df.drop([0], axis=0)
df2.reset_index(inplace=True)
```

```python
# Eliminamos la columna index

del df2['index']
```

```python
# Movemos los valores a la columna que pertencen, ya que estan desordenados
 
for i in range(0,len(df2)):
        if df2.iloc[i][0] != 'Comprobante: Liquidación':
                df2.iloc[i][6] = df2.iloc[i][2]
                df2.iloc[i][5] = df2.iloc[i][1]
                df2.iloc[i][4] = df2.iloc[i][0]
        
```

```python
# Creamos una variable con los nombres de las columnas que queremos, luego asignamos dicha variable a las columnas del DF

names_columns = ['comprobante', 'n_comprobante', 'fecha', 'cliente', 'codigo', 'cantidad', 'descripcion', 'p_unitario', 'iva', 'precio_final']
df2.columns = names_columns
```

```python
# A la columna que tiene las fechas, le quitamos las palabras que estan antes de las fechas (Fecha Comprobante:)

for i in range(0,len(df2)):
        if df2.iloc[i]['comprobante'] == 'Comprobante: Liquidación':
                df2.iloc[i]['fecha'] = df2.iloc[i]['fecha'].strip('Fecha Comprobante: ')
        
```

```python
# Rellenamos con None las filas siguientes en la columna fecha y n_comprobante 

for i in range(0,len(df2)):
        if df2.iloc[i]['comprobante'] != 'Comprobante: Liquidación':
            df2.iloc[i]['fecha'] = None
            df2.iloc[i]['n_comprobante'] = None
```

```python
# Rellenamos ahora los registros NA con el valor anterior en fecha y cliente

df2.fecha.fillna(method='ffill', inplace=True)
df2.cliente.fillna(method='ffill', inplace=True)
```

```python
# Rellenamos ahora los registros NA con el valor anterior en comprobante y numero de comprobante

df2.comprobante.fillna(method='ffill', inplace=True)
df2.n_comprobante.fillna(method='ffill', inplace=True)
```

```python
# Comprobamos si la fecha se ha cambiado a formato datetime

df2.fecha.dtype
```

```python
# Eliminamos la columna comprobante

df2.drop('comprobante', axis = 1, inplace = True)
```

```python
# Eliminamos los na
# 
df2.dropna(axis = 0, inplace = True)
```

```python
# Eliminamos las primeras filas de cada comprobante, ya que repite informacion de los nombres de las columnas

df2 = df2[df2['cantidad'] != 'Cantidad']
```

```python
# Ordenamos el dataframe por numero de comprobante
 
df2.sort_values(by= 'n_comprobante', inplace=True)
```

```python
# Convertimos los tipos de datos

df2['fecha'] =  pd.to_datetime(df2['fecha']).dt.strftime('%d/%m/%Y')
df2['cliente']=df2['cliente'].astype('category')
df2['codigo']=df2['codigo'].astype('int')
df2['cantidad']=df2['cantidad'].astype('int')
df2['p_unitario']=df2['p_unitario'].astype('float')
df2['iva']=df2['iva'].astype('float')
df2['precio_final']=df2['precio_final'].astype('float')
```
