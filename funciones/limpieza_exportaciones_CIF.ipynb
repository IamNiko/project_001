```python
# Pasamos la columna final a tipo entero

df['final'] = df['final'].astype('int')
```

```python
# Definimos la variable rubro como categorica

df['rubro'] = df['rubro'].astype('category')
```

```python
# Imprimimos los nombres de las categorias:

df.rubro.cat.categories
```

```python
# Corregimos los nombres de los rubros:

df.replace('Cigarrillos (2)','cigarrillos', inplace= True)
df.replace('Generico (1)','generico', inplace= True)
df.replace('Golosinas (2)','golosinas', inplace= True)
df.replace('Helados (1.504)','helados', inplace= True)
df.replace('Krachitos','kranchitos', inplace= True)
df.replace('Golosinas (3)','golosinas', inplace= True)
df.replace('9327800004825','encendedores', inplace= True)
df.replace(['almacen (15)', '7790150320266'], 'almacen', inplace= True)
```

```python
# Rellenamos los NA de la variable rubro con la categoria almacen

df.rubro.fillna('almacen', inplace = True)
```

```python
# Verificamos los que no haya nulos en el DF

df.isnull().sum()
```

```python
# Contamos la cantidad de categorias:

cate = df.rubro.value_counts()
len(cate)
```

```python
# Creamos un index, para ello sacamos el len del DF primero:

w = len(df)
df.insert(0, column= 'INDEX', value= range(0,w,1))
df.set_index('INDEX', inplace= True)
```

```python
# Agregamos las variables MARGEN, IVA, COSTO, NETO, GANANCIA, para el rubro cigarrillos ponemos un margen distinto

for rub in df['rubro']:
    if rub != 'cigarrillos':
        df['margen'] = 55
        df['IVA'] = 0.21
        df['costo'] = df['final'].round() / 1.55
        df['neto'] = df['final'].round() / (1 + df['IVA'])
        df['ganancia'] = df['final'].round() - df['costo'].round()

    else:
        df['margen'] = 10
        df['costo'] = df['final'] / 1.1
        df['neto'] = df['final'] - df['costo']
        df['ganancia'] = df['final'] - df['costo']
        
df.round(2)
```
