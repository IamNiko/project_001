Importamos las librerias.

```python
# Importamos las librerias a utilizar

import pandas as pd
import numpy as np
from janitor import clean_names
```

```python
# Creamos la variable con el nombre del archivo

nombre_archivo = '2022total.xlsx'
```

```python
# Leemos el archivo y Creamos la variable df donde guardamos el DF:

df = pd.read_excel(f'data/{nombre_archivo}', na_values= 'NN')
```

```python
# Creamos una variable nueva llamada DF2 donde guardamos sobre una copia de DF, a su vez redondeamos a dos digitos los decimales:

df2 = df.copy()
df2 = df2.round(2)
df2
```

```python
# Sacamos en consola los nombres de las columnas del DF:

df2.columns
```

```python
# Creamos una variable llamada nombres nuevos con los nombres y Renombramos las columnas:

nombres_nuevos = ['rubro', 'articulo', '1','2','cantidad', 'final', 'costo', 'neto', 'margen']

df2.columns = nombres_nuevos
```

```python
# Sacamos la informacion del DF:

df2.info()
```

```python
# Sacamos los nulos del DF por variable:

df2.isna().sum().sort_values(ascending = False)
```

```python
# Usamos la funcion clean_names para limpiar los nombres de las variables:

clean_names(df2)
```

```python
# Rellenamos los nulos en la variable RUBRO, con el metodo FFILL ya que todos los registros pertenecen al mismo rubro:

df2.rubro.fillna(method='ffill', inplace= True)
```

```python
# Creamos las funciones que nos permitiran corregir los registros donde aparecen los totales por rubro, ya que los valores
# estan desordenados respecto al resto de los registros.

def correccion_datos(i):
    for i in range(len(df2)):
        df2.index.values[i]
        if df2.iloc[i,0].startswith('Total Rubro '):
            df2.drop
```

Ejecutamos las funciones de correccion una por una.

```python
df2.apply(correccion_datos)
```

```python
# Eliminamos el ultimo registro ya que no es util:

w = len(df2)
e = w -1
df2.drop(index = df2.index[e], inplace=True)
```

```python
# Eliminamos las columnas 1 y 2 que solo contienen nulos:

df2.drop(columns=['1','2'], inplace= True)
```

```python
# Eliminamos el ultimo registro que no es de utilidad:

w = len(df2)
e = w -1
df2.drop(index= df2.index[e],inplace= True)
```

```python
# Eliminamos los registros que tienen nulos en la variable Neto:

df2.drop(df2[df2.neto.isnull()].index, inplace = True)
```

```python
# Comprobamos si aun quedan nulos:

df2.isna().sum().sort_values(ascending = False)
```

```python
df2['final'] = df2['final'].astype('int')
```

```python
# Definimos a rubro como categorias

df2['rubro'] = df2['rubro'].astype('category')
```

```python
# Imprimimos los nombres de las categorias:

df2.rubro.cat.categories
```

```python
# Corregimos los nombres de los rubros:

df2.replace('Cigarrillos (2)','cigarrillos', inplace= True)
df2.replace('Generico (1)','generico', inplace= True)
df2.replace('Golosinas (2)','golosinas', inplace= True)
df2.replace('Helados (1.504)','helados', inplace= True)
df2.replace('Krachitos','kranchitos', inplace= True)
df2.replace('Golosinas (3)','golosinas', inplace= True)
df2.replace('9327800004825','encendedores', inplace= True)
df2.replace(['almacen (15)', '7790150320266'], 'almacen', inplace= True)
```

```python
# Contamos la cantidad de categorias:

cate = df2.rubro.value_counts()
len(cate)
```

```python
# Creamos un index, para ello sacamos el len del DF primero:

w = len(df2)
df2.insert(0, column= 'INDEX', value= range(0,w,1))
df2.set_index('INDEX', inplace= True)
```

```python
for rub in df2['rubro']:
    if rub != 'cigarrillos':
        df2['margen'] = 55
        df2['IVA'] = 0.21
        df2['costo'] = df2['final'].round() / 1.55
        df2['neto'] = df2['final'].round() / (1 + df2['IVA'])
        df2['ganancia'] = df2['final'].round() - df2['costo'].round()

    else:
        df2['margen'] = 10
        df2['costo'] = df2['final'] / 1.55
        df2['neto'] = df2['final'] - df2['costo']
        df2['ganancia'] = df2['final'] - df2['costo']
        
df2.round(2)
```

```python
df2['final']=df2['final'].astype('int64')
```

Exportamos el DF en un archivo CSV y un EXCEL.

```python
# Creamos una variable con el nuevo nombre para guardar el DF transformado

nombre_nuevo = nombre_archivo.strip('.xlsx')
nombre_transformado = f'{nombre_nuevo}_transformado'
```

```python
# Guardamos el archivo como csv y excel, con el mismo nombre original mas la palabra transformado

df2.to_csv(f'{nombre_transformado}.csv')
df2.to_excel(f'{nombre_transformado}.xlsx')
```
